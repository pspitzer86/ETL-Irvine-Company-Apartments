```python
from splinter import Browser
from bs4 import BeautifulSoup as bs
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager

driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
cities_population_url = "https://www.california-demographics.com/cities_by_population"
browser.visit(cities_population_url)
```

```python
html = browser.html
soup = bs(html, 'html.parser')
```

```python
import requests
r = requests.get(cities_population_url)
r.text
```

```python
soup = bs(r.text,'html.parser')
print(soup.prettify())
```

```python
results = soup.find('table', class_="ranklist")
print(results)
```

```python
import re
import pandas as pd
```

```python
body1 = results.find_all("tr")
head = body1[0]
body_rows = body1[1:]
headings = []
for item in head.find_all("th"):
    item = item.text.rstrip("/n")
    headings.append(item)
    print(headings)
    
```

```python
all_rows = [] 
for row_num in range(len(body_rows)): 
    row = [] 
    for row_item in body_rows[row_num].find_all("td"): 
        aa = re.sub("(\xa0)|(\n)|,","",row_item.text)       
        row.append(aa)    
        all_rows.append(row)
```

```python
df = pd.DataFrame(data=all_rows, columns=headings)
new_df = df.drop_duplicates()
city_df = new_df.reset_index()
city_df = city_df.drop(columns=["index"])
city_df
```

```python
base_url = "https://www.california-demographics.com/"
each_city_url = []
for link in results.find_all('a'):       
        city_link_url = base_url + link.get('href')
        each_city_url.append(city_link_url)
print(each_city_url)
```

```python
each_city_median = []
for each_city in each_city_url:
    browser.visit(each_city)
    html = browser.html
    soup = bs(html, 'html.parser')
    re = requests.get(each_city)
    soup = bs(re.text,'html.parser')
    median = soup.find('div', class_="col summary_block")
    each_city_median.append(median)
    print(median)
```

```python
each_city_income = []
median_city_age = []
for median_income in each_city_median:
    text = median_income.find_all('p')[2].text
    ds_pos = text.index('$')
    point_pos = text.index('.')
    income = int(text[ds_pos:point_pos].replace('$','').replace(',',''))
    median_age_text = median_income.find_all('p')[3].text
    age_locate_pos =median_age_text.split()
    median_age_pos = age_locate_pos[8]
    each_city_income.append(income)
    median_city_age.append(median_age_pos)
```

```python
median_city_age_df = pd.DataFrame(data=median_city_age, columns=["median_age"])
median_city_age_df
```

```python
each_city_income_df = pd.DataFrame(data=each_city_income, columns=["median_income"])
each_city_income_df
```

```python
new_each_city_df = city_df.iloc[:100]
new_each_city_df
```

```python
general_info_df = pd.concat([new_each_city_df,each_city_income_df,median_city_age_df], axis=1)
general_info_df
```

```python
total_df = pd.merge(city_df, general_info_df, how='left', on=["Rank","City","Populatio"])
total_df = total_df.rename(columns={'Populatio':'Population'})
total_df
```

```python
total_df = total_df.dropna(how="any")
total_df
```

```python
total_df = total_df.loc[(total_df["median_age"] != "years") & (total_df["median_age"] != "is")]
total_df.head()
```

```python
city_list = total_df["City"].to_list()
pop_list = total_df["Population"].to_list()
income_list = total_df["median_income"].to_list()
age_list = total_df["median_age"].to_list()
```

```python
from sqlalchemy import create_engine, insert
from config import password
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
```

```python
engine = create_engine(f'postgresql://postgres:{password}@localhost:5432/ETL_IrvineCoApts_db')
connection = engine.connect()
```

```python
Base = automap_base()
Base.prepare(engine, reflect=True)
```

```python
Cities = Base.classes.cities
```

```python
session = Session(connection)
```

```python
i = 0
for name in city_list:

    results = session.query(Cities.city_id).filter(Cities.city_name == name).first()
    if results == None:
        new_city = Cities(city_name = name, cost_of_living = 0.0, population = pop_list[i], \
                          median_income = income_list[i], median_age = age_list[i])
        session.add(new_city)
    else:
        city_id = results[0]
        session.query(Cities)\
                      .filter(Cities.city_id == city_id)\
                      .update({Cities.population: pop_list[i], \
                               Cities.median_income: income_list[i], \
                               Cities.median_age: age_list[i]})
    i += 1
        
        
session.commit()
```
